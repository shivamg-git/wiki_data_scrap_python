# Scrapping the Links of state from page.

```python
file = open("./States_and_union_territories_of_India.html","r")
content = file.read()
soup = BeautifulSoup(content)
tables = soup.findAll("table",{'class': "wikitable"})
table1 = tables[0].findAll("th",{'scope':"row"})
table2 = tables[1].findAll("th",{'scope':"row"})

state_list=[]
for t in table1:
    for a in t.findAll('a',href=True):
        state_list.append("https://en.wikipedia.org/"+a['href'])

for t in table2:
    for a in t.findAll('a',href=True):
        state_list.append("https://en.wikipedia.org/"+a['href'])

state_list
```

# Loading Pages from Downloaded Page

```python
states=[]
for url in state_list:
    state={}
    title = url.split("/")[-1]
    state['title']=title
    state['url']  = url
    file = open(title+".html",'r')
    content = file.read()
    soup = BeautifulSoup(content)
    state['content']=soup
    print(soup)
    file.close()
    states.append(state)
```

# finding div that conten data in page

```python
for state in states:
    religion_div = None
    mydivs = state['content'].findAll("div", {"class": "thumbcaption"})
    for div in mydivs:
        legend_div = div.findAll("div", {"class": "legend"})
        if(len(legend_div) !=0):
            pp = div.findAll("p")
            t = pp[0].text
            if("Religion" in t):
                religion_div = div
    state['religion_div']=religion_div

for state in states:
    print(state['religion_div'])
    print("-"*30)
```

# Extracting Content from div

```python
for state in states:
    if(state['religion_div']):
        religion_div=state['religion_div']
        rel_divs = religion_div.findAll("div", {"class": "legend"})
        rel_dict=[]
        for rel in rel_divs:
            rel_dict.append(rel.text)
        state['rel_dict'] = rel_dict
for state in states:
    try:
        print(state['rel_dict'])
        print("-"*30)
    except:
        print("NONE")
```

# making key value pair for data

```python
rel_list=set()
for state in states:
    if(state['religion_div']):
        per={}
        for s in state['rel_dict']:
            x = s.split()
            x = [x[0].strip(),x[-1].strip()]
            if(x[0]=="Not" or x[0]=="No" or x[0]=="not"):
                x[0]="NONE"
            per[x[0]]=float(x[-1][1:-2])
            rel_list.add(x[0])
        state["data"]=per
for state in states:
    try:
        print(state['data'])
        print("-"*30)
    except:
        print("NONE")
```
